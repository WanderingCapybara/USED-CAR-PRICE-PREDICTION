## **Abnormal Values Handling**

**In the previous EDA stage, there are rows that had extreme and unreasonable values. So, in this data cleaning stage these rows will be deleted.**

```python
# Dropping abnormal 'odometer' values
i_min_odo = dfclean[((dfclean.odometer <= 100))].index
dfclean = dfclean.drop(i_min_odo)
```

```python
# Dropping abnormal 'sellingprice' values
i_min_price = dfclean[((dfclean.sellingprice == 1))].index
dfclean = dfclean.drop(i_min_price)
```

```python
dfclean.describe()
```

## **Missing Values Handling**

**Some features contain missing values. At this stage this problem will be handled in several ways, depending on the percentage of missing values in the total data. The missing values below 1% will be dropped. Otherwise, they will be replaced with mode for categorical features and mean for numerical features.**

```python
# Checking the percentage of missing values in features
null_percentage= (dfclean.isnull().sum() / len(dfclean)) * 100
print(null_percentage)
```

```python
# Defining some features to be processed
mode_fill = ['make', 'model', 'trim', 'body', 'transmission']
to_drop = ['odometer', 'color', 'interior']
```

```python
# Filling out null values with mode
for col in mode_fill:
    dfclean[col] = dfclean[col].fillna(dfclean[col].mode().iloc[0])
```

```python
# Dropping null values
for col in to_drop:
    dfclean = dfclean.dropna(subset=[col])
```

```python
# Filling out null values with mean
dfclean['condition'] = dfclean['condition'].fillna(dfclean['condition'].mean())
```

```python
dfclean.isna().sum()
```

**Before I move on to the next process, remember that `color` and `interior` features contain value that is ''. Therefore, these values will be replaced with mode.**

```python
# Replacing '' values with mode
dfclean['color'] = dfclean['color'].replace({'': dfclean['color'].mode().iloc[0]})
dfclean['interior'] = dfclean['color'].replace({'': dfclean['color'].mode().iloc[0]})
```

## **Feature Encoding**

**This process aims to translate categorical values into numerical ones. The encoding method used is target encoding to avoid the curse of dimensionality. This method is also effective in preserving information about the relationships with other features, especially the mmr feature. The choice of using `mmr` as the target feature is because MMR is considered a standard car price valuation based on several variables. Therefore, I believe that the categorical features to be encoded have a fairly strong relationship with the `mmr` feature.**

```python
# Defining categorical features to be encoded
to_encode = ['make', 'model', 'trim', 'body', 'state', 'color', 'interior', 'seller']
```

```python
# Performing target encoding for categorical features
for col in to_encode:
    encoding_map = dfprep.groupby(col)['mmr'].mean().to_dict()
    dfprep[col + '_encoded'] = dfprep[col].map(encoding_map)
```

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split

# Defining features and target
X= dfmodel.drop('sellingprice', axis= 1)
y= dfmodel['sellingprice']

# Splitting train-test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 0)
```

## **Feature Scaling**

**The feature scalling is conducted so that different features have the same scale. RobustScaler is choosen because it can handle outliers contained in the data.**

```python
from sklearn.preprocessing import RobustScaler

# Creating a RobustScaler object
rs = RobustScaler()

# Features to be normalized
feat_scale = dfmodel.drop('sellingprice', axis= 1).columns

# Fitting the scaler on data
for col in feat_scale:
    scaler = rs.fit(X_train[[col]])
    X_train[col] = scaler.transform(X_train[[col]])
    X_test[col] = scaler.transform(X_test[[col]])
```

## **Model Training**

**XGBoost is tha algorithm that will be used in this proccess. The reason is that XGBoost is a powerful and versatile algorithm that is widely used in machine learning for its ability to deliver excellent predictive performance across various domains and datasets.**

```python
from sklearn.metrics import mean_squared_error, r2_score

# Creating a function to see the evaluation score of a model
def model_training(model, X_train, y_train, X_test, y_test):
    
    # Make predictions on the training and test data
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)

    # Calculate RMSE for train and test data
    rmse_train = mean_squared_error(y_train, train_predictions, squared=False).round(3)
    rmse_test = mean_squared_error(y_test, test_predictions, squared=False).round(3)

    # Calculate R2 score for train and test data
    r2_train = r2_score(y_train, train_predictions).round(3)
    r2_test = r2_score(y_test, test_predictions).round(3)

    # Display the evaluation scores
    print("RMSE (Train):", rmse_train)
    print("RMSE (Test):", rmse_test)
    print("R2_score (Train):", r2_train)
    print("R2_score (Test):", r2_test)
```

```python
from xgboost import XGBRegressor
# Training the model and displaying the evaluation score

xgb = XGBRegressor(random_state= 0)
xgb.fit(X_train, y_train)

model_training(xgb, X_train, y_train, X_test, y_test)
```

**As evident from the results, the model exhibits strong performance, as indicated by a high R-squared (R2) score and low Root Mean Squared Error (RMSE) values. Furthermore, the minimal disparity between the training and test set scores suggests that the model is not overfitting.**

## **Hyperparameter Tuning**

**Even though the result from the initial training is strong, I want to see if I can make the model's performance better by tuning the parameters. Here, I use bayesian optimization to perform the evaluation because of it's better efficiency than grid search or random search**

```python
pip install optuna
```

```python
# Creating function to find the best parameters for the model
import optuna
def objective(trial):
    
    # Defining the hyperparameter search space
    params = {
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 20),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'n_estimators': 100,  # You can set a fixed number of estimators here
        'objective': 'reg:squarederror',
        'random_state': 0
    }

    # Creating and training the XGBoost model
    model = XGBRegressor(**params)
    model.fit(X_train, y_train)

    # Making predictions on the test set
    y_pred = model.predict(X_test)

    # Calculating RMSE as the objective to minimize
    rmse = mean_squared_error(y_test, y_pred, squared=False)

    return rmse

if __name__ == "__main__":
    study = optuna.create_study(direction='minimize')
    study.optimize(objective, n_trials=100)

    # Printing the best hyperparameters and RMSE
    print("Best trial:")
    trial = study.best_trial
    print("  RMSE: {:.3f}".format(trial.value))
    print("  Hyperparameters: ")
    for key, value in trial.params.items():
        print("    {}: {}".format(key, value))
```

## **Model Re-training**

**I will retraing the model by including the parameters provided in the hyperparameters tunning process to improve its performance.**

```python
# Defining the best hyperparameters
best_hyperparameters = {
    'max_depth': 10,
    'min_child_weight': 7,
    'gamma': 0.10643003755782982,
    'colsample_bytree': 0.9993267778438296,
    'learning_rate': 0.2514333608442605,
    'n_estimators': 100,
    'objective': 'reg:squarederror',
    'random_state': 0
}

# Creating and training the XGBoost model with the best hyperparameters
xgb = XGBRegressor(**best_hyperparameters)
xgb.fit(X_train, y_train)

# Evaluating the model
model_training(xgb, X_train, y_train, X_test, y_test)
```

**As can be seen from the results above, including the parameters has led to a slight improvement in the model's performance.**

## **Model Interpretation: SHAP values**

**SHAP values have several main functions in the interpretation of Machine Learning models. Not only they help in understanding the complexity of the model by explaining the contribution of each feature to the predicted results, they help explain the causes of changes in predictions, thereby identifying features that lead to increases or decreases in predictive value. Furthermore, SHAP values enable analysis of each feature's impact on predictions on a case-by-case level, providing insight into how each feature affects outcomes for individual data. As such, SHAP values are an important tool for the interpretation and analysis of Machine Learning models.**

```python
pip install shap
```

### **Summary Plot**

```python
import shap

explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test)
```

```python
# Creating df containing SHAP values
mean_shap_values = shap_values.mean(axis=0)
dfshap = pd.DataFrame(index= X_test.columns, columns= ['Average Shap Value'], data= mean_shap_values)
dfshap['Average Shap Value'].sort_values(ascending= False)
```

**Following is the interpretation of SHAP values for each feature:**

`model_encoded`: SHAP value of approximately 42098.797 indicates that the vehicle model (encoded) has a significant positive impact on price. Certain models tend to be more expensive than others.

`odometer`: SHAP value around -36313.035 indicates that the higher the odometer number (mileage), the lower the price of the vehicle. This is normal because vehicles with lower mileage tend to be more expensive.

`trim_encoded`: SHAP value of approximately 16663.96 indicates that vehicle trim type (encoded) has a significant positive impact on price. Some trim types tend to be more expensive.

`seller_encoded`: SHAP value of around 11464.917 indicates that the type of vehicle seller (encoded) has a significant positive impact on price. Some types of sellers tend to offer higher prices

`body_encoded`: A SHAP value of around 2023.1923 indicates that the vehicle body type (encoded) has a smaller positive effect on price.

`condition`: SHAP value of around 1684.848 indicates that higher condition values tend to drive improvements in price predictions. In other words, when the condition of the vehicle is better, prices tend to be higher.

`make_encoded`: SHAP value of around 50.771217 indicates that the make of the (encoded) vehicle makes a relatively small contribution to the change in price.
